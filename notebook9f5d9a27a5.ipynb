```python
#include <stdio.h>

int main() {
    int n, first = 0, second = 1, next, i;

    // Ask the user for the number of terms
    printf("Enter the number of terms for the Fibonacci series: ");
    scanf("%d", &n);

    printf("Fibonacci Series: ");

    for (i = 0; i < n; i++) {
        if (i <= 1) {
            next = i;  // The first two terms are 0 and 1
        } else {
            next = first + second;
            first = second;
            second = next;
        }
        printf("%d ", next);
    }

    return 0;
}
```
