```python
#include <stdio.h>
#include <pthread.h>

// Function to print "Hello"
void* print_hello(void* arg) {
    printf("Hello\n");
    return NULL;
}

// Function to print "World"
void* print_world(void* arg) {
    printf("World\n");
    return NULL;
}

int main() {
    pthread_t thread1, thread2;
    
    // Creating the first thread to print "Hello"
    if (pthread_create(&thread1, NULL, print_hello, NULL) != 0) {
        printf("Error creating thread1\n");
        return 1;
    }

    // Creating the second thread to print "World"
    if (pthread_create(&thread2, NULL, print_world, NULL) != 0) {
        printf("Error creating thread2\n");
        return 1;
    }

    // Waiting for threads to finish
    pthread_join(thread1, NULL);
    pthread_join(thread2, NULL);

    return 0;
}
```
